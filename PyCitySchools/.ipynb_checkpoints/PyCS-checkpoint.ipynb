# SUMMARY:
  This script is used to analyse the school data of a local goverment area to helpp the school board and the mayor to make  decisions on budgets and priorities. The data are summarised as:
  (1) for the whole local goverment area; then by 
  (2) schools; and their 
  (3) students performance based on their scores on maths, reading, and overall - per school and per year level;
  (4) spenditures; 
  (5) school size; and 
  (6) school type

# TRENDS:
  * One of the most apparent trends is we could see that the performance of the students are affected by the school size. As seen from the **size_summary** table, all the passing scores and the passing rates decrease as the school size becomes larger. A similar observation can be seen in the school type as well. Indpendent schools seem to have higher performing students than the ones in Goverment schools, which again can be attributed to the student population. **per_school_summary** table shows that Goverment schools have higher number of total students as compared to Independent schools.
  
  
  * In terms of school spending, **spending_summary** table shows that the most optimal budget allocated per student is about S585-630 per student, spending more or less than this range actually shows a decrease in the students performance. Looking at the **per_school_summary** table, we can see that the Independent schools' budgets per student are well within this range. At the same time, it is also worth noting that the Government schools have higher budget per student than the Independent schools.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

# Local Government Area Summary
* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Local Government Area Summary

total_school = school_data_complete['School ID'].value_counts().count()

total_student = school_data_complete['Student ID'].value_counts().count()

total_school_budget = school_data_complete['budget'].unique().sum()

total_ave_maths = school_data_complete['maths_score'].mean()

total_ave_reading = school_data_complete['reading_score'].mean()

pass_maths = school_data_complete.loc[(school_data_complete["maths_score"] >= 50)]
total_pass_maths = pass_maths.value_counts().count()
rate_pass_maths = total_pass_maths/total_student

pass_reading = school_data_complete.loc[(school_data_complete["reading_score"] >= 50)]
total_pass_reading = pass_reading.value_counts().count()
rate_pass_reading = total_pass_reading/total_student

pass_overall = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) 
                                        & (school_data_complete["reading_score"] >= 50)]
total_pass_overall = pass_overall.value_counts().count()
rate_pass_overall = total_pass_overall/total_student

area_summary = pd.DataFrame({"Total Schools":[total_school], "Total Students":[total_student],
                                     "Total Budget":[total_school_budget], "Average Maths Score":[total_ave_maths],
                                     "Average Reading Score":[total_ave_reading], "% Passing Maths":[rate_pass_maths],
                                     "% Passing Reading":[rate_pass_reading] ,"% Overall Passing":[rate_pass_overall]})

area_summary["Total Students"] = area_summary["Total Students"].astype(float).map("{:,.0f}".format)
area_summary["Total Budget"] = area_summary["Total Budget"].astype(float).map("${:,.2f}".format)
area_summary["Average Maths Score"] = area_summary["Average Maths Score"].astype(float).map("{:,.2f}".format)
area_summary["Average Reading Score"] = area_summary["Average Reading Score"].astype(float).map("{:,.2f}".format)
area_summary["% Passing Maths"] = area_summary["% Passing Maths"].astype(float).map("{:,.2%}".format)
area_summary["% Passing Reading"] = area_summary["% Passing Reading"].astype(float).map("{:,.2%}".format)
area_summary["% Overall Passing"] = area_summary["% Overall Passing"].astype(float).map("{:,.2%}".format)

area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Summary

pass_school_maths = school_data_complete.loc[school_data_complete["maths_score"] >= 50]
pass_school_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 50]
pass_school_overall = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]

grouped_maths = pass_school_maths.groupby(['school_name','type'])
reduced_maths = grouped_maths[["student_name"]].count()
reduced_maths

grouped_reading = pass_school_reading.groupby(['school_name','type'])
reduced_reading = grouped_reading[["student_name"]].count()
reduced_reading

grouped_overall = pass_school_overall.groupby(['school_name','type'])
reduced_overall = grouped_overall[["student_name"]].count()
reduced_overall

merge_mr = pd.merge(reduced_maths, reduced_reading, on=["school_name","type"])
merged_grades = pd.merge(merge_mr, reduced_overall, on=["school_name","type"])
merged_grades = merged_grades.rename(columns={"student_name_x":"passed_maths","student_name_y":"passed_reading","student_name":"passed_all"})
merged_grades

grouped_school_data = school_data_complete.groupby(['school_name','type'])
grouped_school_data.head()

scores_school_data = grouped_school_data[["reading_score","maths_score"]].sum()
scores_school_data

count_school_data = grouped_school_data["Student ID"].count()
count_school_data

budget_school_data = grouped_school_data["budget"].unique()
budget_school_data

merge_school_data = pd.merge(scores_school_data, count_school_data, on=["school_name","type"])
merged_school_data = pd.merge(merge_school_data,budget_school_data, on=["school_name","type"])

merged_school_data = merged_school_data.rename(columns={"Student ID":"Total Students","budget":"Total School Budget"})

merged_school_data["Per Student Budget"] = merged_school_data["Total School Budget"]/merged_school_data["Total Students"]
merged_school_data["Average Maths Score"] = merged_school_data["maths_score"]/merged_school_data["Total Students"]
merged_school_data["Average Reading Score"] = merged_school_data["reading_score"]/merged_school_data["Total Students"]

merged_grade_data = pd.merge(merged_school_data, merged_grades, on=["school_name","type"])

merged_grade_data["% Passing Maths"] = merged_grade_data["passed_maths"]/merged_grade_data["Total Students"]
merged_grade_data["% Passing Reading"] = merged_grade_data["passed_reading"]/merged_grade_data["Total Students"]
merged_grade_data["% Overall Passing"] = merged_grade_data["passed_all"]/merged_grade_data["Total Students"]

merged_grade_data = merged_grade_data.reset_index(drop=False)
merged_grade_data = merged_grade_data.rename(columns={"school_name":"School Name","type":"School Type"})

per_school_summary = pd.DataFrame(merged_grade_data[["School Name","School Type","Total Students","Total School Budget","Per Student Budget","Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]]).set_index("School Name")
per_school_summary


per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].astype(float).map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].astype(float).map("${:,.2f}".format)
per_school_summary["Average Maths Score"] = per_school_summary["Average Maths Score"].astype(float).map("{:,.2f}".format)
per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].astype(float).map("{:,.2f}".format)
per_school_summary["% Passing Maths"] = per_school_summary["% Passing Maths"].astype(float).map("{:,.2%}".format)
per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].astype(float).map("{:,.2%}".format)
per_school_summary["% Overall Passing"] = per_school_summary["% Overall Passing"].astype(float).map("{:,.2%}".format)

per_school_summary.index.name = None
per_school_summary
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#Top Performing Schools (By % Overall Passing)

top_schools = per_school_summary.sort_values("% Overall Passing",ascending=False)
top_schools.index.name = None
top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
#Bottom Performing Schools (By % Overall Passing)

bottom_schools = per_school_summary.sort_values("% Overall Passing",ascending=True)
bottom_schools.index.name = None
bottom_schools.head(5)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Maths Scores by Year

y9_maths = school_data_complete.loc[school_data_complete["year"] == 9,:].groupby(['school_name'])[["school_name","maths_score"]].mean()
y9_maths = y9_maths.rename(columns={"maths_score":"Year 9"})
y10_maths = school_data_complete.loc[school_data_complete["year"] == 10,:].groupby(['school_name'])[["school_name","maths_score"]].mean()
y10_maths = y10_maths.rename(columns={"maths_score":"Year 10"})
y11_maths = school_data_complete.loc[school_data_complete["year"] == 11,:].groupby(['school_name'])[["school_name","maths_score"]].mean()
y11_maths = y11_maths.rename(columns={"maths_score":"Year 11"})
y12_maths = school_data_complete.loc[school_data_complete["year"] == 12,:].groupby(['school_name'])[["school_name","maths_score"]].mean()
y12_maths = y12_maths.rename(columns={"maths_score":"Year 12"})

maths_scores_by_year = pd.concat([y9_maths,y10_maths,y11_maths,y12_maths],axis=1)
maths_scores_by_year.index.name = None

maths_scores_by_year["Year 9"] = maths_scores_by_year["Year 9"].astype(float).map("{:,.2f}".format)
maths_scores_by_year["Year 10"] = maths_scores_by_year["Year 10"].astype(float).map("{:,.2f}".format)
maths_scores_by_year["Year 11"] = maths_scores_by_year["Year 11"].astype(float).map("{:,.2f}".format)
maths_scores_by_year["Year 12"] = maths_scores_by_year["Year 12"].astype(float).map("{:,.2f}".format)

maths_scores_by_year
```

## Reading Score by Year

* Perform the same operations as above for reading scores

```python
#Reading Score by Year

y9_reading = school_data_complete.loc[school_data_complete["year"] == 9,:].groupby(['school_name'])[["school_name","reading_score"]].mean()
y9_reading = y9_reading.rename(columns={"reading_score":"Year 9"})
y10_reading = school_data_complete.loc[school_data_complete["year"] == 10,:].groupby(['school_name'])[["school_name","reading_score"]].mean()
y10_reading = y10_reading.rename(columns={"reading_score":"Year 10"})
y11_reading = school_data_complete.loc[school_data_complete["year"] == 11,:].groupby(['school_name'])[["school_name","reading_score"]].mean()
y11_reading = y11_reading.rename(columns={"reading_score":"Year 11"})
y12_reading = school_data_complete.loc[school_data_complete["year"] == 12,:].groupby(['school_name'])[["school_name","reading_score"]].mean()
y12_reading = y12_reading.rename(columns={"reading_score":"Year 12"})

reading_scores_by_year = pd.concat([y9_reading,y10_reading,y11_reading,y12_reading],axis=1)
reading_scores_by_year.index.name = None

reading_scores_by_year["Year 9"] = reading_scores_by_year["Year 9"].astype(float).map("{:,.2f}".format)
reading_scores_by_year["Year 10"] = reading_scores_by_year["Year 10"].astype(float).map("{:,.2f}".format)
reading_scores_by_year["Year 11"] = reading_scores_by_year["Year 11"].astype(float).map("{:,.2f}".format)
reading_scores_by_year["Year 12"] = reading_scores_by_year["Year 12"].astype(float).map("{:,.2f}".format)

reading_scores_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Scores by School Spending

spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

merged_grade_data["Per Student Budget"] = merged_grade_data["Per Student Budget"].astype('int64')
merged_grade_data["Spending Ranges (Per Student)"] = pd.cut(merged_grade_data['Per Student Budget'], bins = spending_bins, labels = labels)

spending_maths_scores = merged_grade_data.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_scores = merged_grade_data.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = merged_grade_data.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = merged_grade_data.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = merged_grade_data.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

spending_summary1 = pd.merge(spending_maths_scores, spending_reading_scores, how="left",on="Spending Ranges (Per Student)")
spending_summary2 = pd.merge(spending_summary1, spending_passing_maths, how="left",on="Spending Ranges (Per Student)")
spending_summary3 = pd.merge(spending_summary2, spending_passing_reading, how="left",on="Spending Ranges (Per Student)")
spending_summary = pd.merge(spending_summary3, overall_passing_spending, how="left",on="Spending Ranges (Per Student)")

spending_summary["Average Maths Score"] = spending_summary["Average Maths Score"].astype(float).map("{:,.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].astype(float).map("{:,.2f}".format)
spending_summary["% Passing Maths"] = spending_summary["% Passing Maths"].astype(float).map("{:,.2%}".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].astype(float).map("{:,.2%}".format)
spending_summary["% Overall Passing"] = spending_summary["% Overall Passing"].astype(float).map("{:,.2%}".format)

spending_summary

# spend_school_summary = merged_grade_data.groupby(["Spending Ranges (Per Student)"])
# spending_summary = spend_school_summary[["Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]].mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Scores by School Size

size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

merged_grade_data["School Size"] = pd.cut(merged_grade_data['Total Students'], bins = size_bins, labels = labels)

size_school_summary = merged_grade_data.groupby(["School Size"])
size_summary = size_school_summary[["Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]].mean()

size_summary["Average Maths Score"] = size_summary["Average Maths Score"].astype(float).map("{:,.2f}".format)
size_summary["Average Reading Score"] = size_summary["Average Reading Score"].astype(float).map("{:,.2f}".format)
size_summary["% Passing Maths"] = size_summary["% Passing Maths"].astype(float).map("{:,.2%}".format)
size_summary["% Passing Reading"] = size_summary["% Passing Reading"].astype(float).map("{:,.2%}".format)
size_summary["% Overall Passing"] = size_summary["% Overall Passing"].astype(float).map("{:,.2%}".format)

size_summary
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
#Scores by School Type

type_school_summary = merged_grade_data.groupby(["School Type"])
type_summary = type_school_summary[["Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]].mean()

type_summary["Average Maths Score"] = type_summary["Average Maths Score"].astype(float).map("{:,.2f}".format)
type_summary["Average Reading Score"] = type_summary["Average Reading Score"].astype(float).map("{:,.2f}".format)
type_summary["% Passing Maths"] = type_summary["% Passing Maths"].astype(float).map("{:,.2%}".format)
type_summary["% Passing Reading"] = type_summary["% Passing Reading"].astype(float).map("{:,.2%}".format)
type_summary["% Overall Passing"] = type_summary["% Overall Passing"].astype(float).map("{:,.2%}".format)

type_summary
```
